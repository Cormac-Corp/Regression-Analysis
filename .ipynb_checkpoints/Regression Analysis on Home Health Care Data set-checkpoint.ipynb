```python
"""
    Author: 
        Paul Lee
        
        
    Objective:
        Perform Regression analysis on the Home Health Care Data set so that I could eventually conduct machine learning algorithms on it.
"""

import pandas              as pd
import numpy               as np
import matplotlib.pyplot   as plt
from dfply                 import *
```

```python
"""
    Read in the data into a pandas dataframe
"""
data_dir = "C:\\Users\\plee\\Desktop\\Main\\data\\HOME_HEALTH_CARE_AGENCIES\\"
data_set = "Home_Health_Care_Agencies.csv"
df = pd.read_csv(data_dir+data_set)
```

```python
"""
    Do an initial heatmap view of the data correlations
"""
df.corr(method = 'spearman').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
"""
    From the initial viewing, we can see the the first 4 columns are not going to be of any use if we're trying to predict the Quality Star Rating. Therefore we will go ahead and drop those 4 columns
"""
df = df.drop(['CMS Certification Number (CCN)*','Zip','Phone','Offers Nursing Care Services'], axis=1)
```

```python
"""
    A more clear view of what's going on
"""
df.corr(method = 'spearman').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
'''
Since this is rudimentary analysis, we'll be looking at 5 features to predict the outcome for the Quality of patient star rating
'''

df_parsed = df[['Quality of Patient Care Star Rating',
         'How often patients got better at walking or moving around',
         'How often patients got better at getting in and out of bed',
         'How often patients got better at bathing',
         'How often patients had less pain when moving around',
         "How often patients' breathing improved"]]

df_parsed = df_parsed.dropna(axis=0)

y = df_parsed['Quality of Patient Care Star Rating']
X = df_parsed[['How often patients got better at walking or moving around',
         'How often patients got better at getting in and out of bed',
         'How often patients got better at bathing',
         'How often patients had less pain when moving around',
         "How often patients' breathing improved",]]
```

```python
'''
 Perform linear regression using the sklearn library
'''

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import metrics


X_train, X_test, Y_train, Y_test= train_test_split(X,y, test_size=0.3)
regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
pred = regr.predict(X_test)
```

```python
'''
What's the accuracy of our model?
'''
accuracy = metrics.r2_score(Y_test,pred)
print(accuracy)
```

```python
"""
    Okay, so our initial results aren't too bad, but we want to run it 100 times to see if what we got wasn't by luck.
"""
for x in range(100):
    X_train, X_test, Y_train, Y_test= train_test_split(X,y, test_size=0.2)
    regr = linear_model.LinearRegression()
    regr.fit(X_train, Y_train)
    pred = regr.predict(X_test)
    accuracy = metrics.r2_score(Y_test,pred)
    print(accuracy)
```
